#FIXING AND CLEANING

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
df['InvoiceDate'].dtype
```

```python
df['Price_clean'] = (
    df['Price']
    .astype(str)
    .str.replace(r'[^0-9.]', '', regex=True)
    .astype(float)
)
```

```python
df.head(5)
```

```python
df['Revenue_clean'] = df['Quantity'] * df['Price_clean']
```

```python
df[['Quantity', 'Price_clean', 'Revenue_clean']].head()
```

#FILTERING VALID SALES DATA

```python
sales_df = df[
    (df['Transaction type'] == 'Sale') &
    (df['Customer ID'].notna())
].copy()
```

```python
sales_df.info()
```

#RFM Analysis

```python
snapshot_date = sales_df['InvoiceDate'].max() + pd.Timedelta(days=1)
```

```python
rfm = sales_df.groupby('Customer ID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'Invoice': 'nunique',
    'Revenue_clean': 'sum'
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
```

```python
rfm['F_bin'] = pd.qcut(
    rfm['Frequency'],
    q=4,
    duplicates='drop'
)
```

```python
rfm['F_score'] = rfm['F_bin'].cat.codes + 1
```

```python
rfm['R_bin'] = pd.qcut(
    rfm['Recency'],
    q=4,
    duplicates='drop'
)

rfm['R_score'] = rfm['R_bin'].cat.codes + 1
rfm['R_score'] = rfm['R_score'].max() - rfm['R_score'] + 1
```

```python
rfm['M_bin'] = pd.qcut(
    rfm['Monetary'],
    q=4,
    duplicates='drop'
)

rfm['M_score'] = rfm['M_bin'].cat.codes + 1
```

```python
rfm['RFM_score'] = rfm[['R_score','F_score','M_score']].sum(axis=1)
```

```python
rfm[['Recency','Frequency','Monetary','R_score','F_score','M_score','RFM_score']].head()
rfm['RFM_score'].describe()
```

```python
def segment(score):
    if score >= 10:
        return 'Champions'
    elif score >= 7:
        return 'Loyal'
    elif score >= 5:
        return 'At Risk'
    else:
        return 'Lost'

rfm['Segment'] = rfm['RFM_score'].apply(segment)
```

```python
rfm['Churn_Risk'] = (rfm['Recency'] > 90).astype(int)
```

```python
rfm
```

#Customer Lifetime Value (CLV) Estimation

```python
rfm['Avg_Order_Value'] = rfm['Monetary'] / rfm['Frequency']
rfm['Purchase_Frequency'] = rfm['Frequency']

rfm['CLV'] = rfm['Avg_Order_Value'] * rfm['Purchase_Frequency']
```

```python
rfm['CLV']
```

#COHORT ANALYSIS

```python
sales_df['OrderMonth'] = sales_df['InvoiceDate'].dt.to_period('M')
sales_df['CohortMonth'] = sales_df.groupby('Customer ID')['InvoiceDate'].transform('min').dt.to_period('M')

cohort = sales_df.groupby(['CohortMonth','OrderMonth'])['Customer ID'].nunique().reset_index()

cohort['CohortIndex'] = (
    cohort['OrderMonth'].astype(int) -
    cohort['CohortMonth'].astype(int)
)
```

#BEHAVIOR ANALYSIS

```python
return_rate_customer = return_rate_customer.rename(
    columns={'Customer ID': 'CustomerID'}
)
```

```python
rfm = rfm.merge(return_rate_customer, on='CustomerID', how='left')
rfm['Return_Value'] = rfm['Return_Value'].fillna(0)
```

```python
rfm[['CustomerID', 'Return_Value']].head()
```

```python
rfm['Return_Value'].describe()
```

```python
rfm.groupby('Segment')['Return_Value'].mean().sort_values(ascending=False)
```

```python
rfm_sorted = rfm.sort_values('Monetary', ascending=False)
rfm_sorted['CumRevenue'] = rfm_sorted['Monetary'].cumsum()
rfm_sorted['CumRevenuePct'] = rfm_sorted['CumRevenue'] / rfm_sorted['Monetary'].sum()
```

```python
rfm_sorted.head()
```

```python


rfm_final = rfm.copy()

rfm_final.to_csv(
    "rfm_customer_analytics_final.csv",
    index=False
)

print("rfm_customer_analytics_final.csv saved")
```

```python


segment_summary = (
    rfm_final
    .groupby('Segment')
    .agg(
        Total_Customers=('CustomerID', 'count'),
        Avg_Recency=('Recency', 'mean'),
        Avg_Frequency=('Frequency', 'mean'),
        Avg_Monetary=('Monetary', 'mean'),
        Avg_CLV=('CLV', 'mean'),
        Avg_Return_Value=('Return_Value', 'mean'),
        Churn_Rate=('Churn_Risk', 'mean')
    )
    .reset_index()
)

segment_summary.to_csv(
    "segment_level_summary.csv",
    index=False
)

print(" segment_level_summary.csv saved")
```

```python


pareto_df = rfm_sorted[['CustomerID', 'Monetary', 'CumRevenuePct']]

pareto_df.to_csv(
    "pareto_customer_revenue.csv",
    index=False
)

print(" pareto_customer_revenue.csv saved")
```
